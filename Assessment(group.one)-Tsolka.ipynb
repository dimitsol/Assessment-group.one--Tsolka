A. Example from the problem description

```python
# Direction mappings
DIRS = {
    '^': (-1, 0),  # move up
    '>': (0, 1),   # move right
    'v': (1, 0),   # move down
    '<': (0, -1)   # move left
}

# Mapping to handle turning right (90° clockwise)
RIGHT = {
    '^': '>',  # up → right
    '>': 'v',  # right → down
    'v': '<',  # down → left
    '<': '^'   # left → up
}


def solve(text):

    # Convert the text input into a 2D grid (list of lists)
    grid = [list(line) for line in text.strip().splitlines()]
    # Number of rows and columns
    R, C = len(grid), len(grid[0])  

    # Find the starting position and direction
    for r in range(R):
        for c in range(C):
            if grid[r][c] in DIRS:
                d = grid[r][c]
                # Start cell is empty for movement
                grid[r][c] = '.'  
                start = (r, c, d)
                break
        else:
            continue
        break

    # Initialize current position and direction
    r, c, d = start

    # Store all visited positions as (row, col) 
    visited = {(r, c)}

    # Simulate movement until the guard leaves the map 
    while True:
        # Direction vector
        dr, dc = DIRS[d]   
        # Next position  
        nr, nc = r + dr, c + dc  

        # If the next move is outside the grid, stop the simulation
        if not (0 <= nr < R and 0 <= nc < C):
            break

        # If the next cell is blocked, turn right
        if grid[nr][nc] == '#':
            d = RIGHT[d]
        else:
            # Move forward 
            r, c = nr, nc
            visited.add((r, c))

    # Return the total number of unique cells visited
    return len(visited)


example = """\
....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#..."""

print("Distinct positions:", solve(example))
```

```python
# Direction mappings
DIRS = {
    '^': (-1, 0),  # move up
    '>': (0, 1),   # move right
    'v': (1, 0),   # move down
    '<': (0, -1)   # move left
}

# Mapping to handle turning right (90° clockwise)
RIGHT = {
    '^': '>',  # up → right
    '>': 'v',  # right → down
    'v': '<',  # down → left
    '<': '^'   # left → up
}


def solve(text):
    # Convert the text input into a 2D grid (list of lists)
    grid = [list(line) for line in text.strip().splitlines()]
    # Number of rows and columns
    rows, cols = len(grid), len(grid[0])

    # Find the starting position and direction
    start = None
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] in DIRS:
                # Start (row, column, direction)
                start = (r, c, grid[r][c])  
                # Replace the guard symbol with empty space
                grid[r][c] = '.'  
                break
        if start:
            break

    # This function simulates the guard's patrol
    def simulate(block=None):
        r, c, d = start
        # Seen (row, col, direction) states
        seen = set()  

        while True:
            # If we visit the same position and direction again → infinite loop
            if (r, c, d) in seen:
                return True
            seen.add((r, c, d))

            # Calculate the next cell the guard wants to move to
            dr, dc = DIRS[d]
            nr, nc = r + dr, c + dc

            # If the next position is outside the grid, guard leaves the map
            if not (0 <= nr < rows and 0 <= nc < cols):
                return False

            # If the next cell is blocked (by # or our new wall), turn right
            if grid[nr][nc] == '#' or (nr, nc) == block:
                d = RIGHT[d]
            else:
                # Move forward
                r, c = nr, nc

    # Placing a new wall in every empty spot and count how many cause loops
    sr, sc, _ = start
    total = 0
    for r in range(rows):
        for c in range(cols):
            # Only consider empty cells ('.')
            if grid[r][c] != '.':
                continue
            # Skip the starting position (the guard is there)
            if (r, c) == (sr, sc):
                continue
            # Run the simulation with a wall placed at (r, c)
            if simulate((r, c)):
                total += 1

    # Return the number of positions that cause looping
    return total

example = """\
....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#..."""

print("Distinct positions:", solve(example))
```
